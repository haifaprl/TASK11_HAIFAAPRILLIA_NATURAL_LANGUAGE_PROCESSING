```python
# Install library yang dibutuhkan
!pip install pandas scikit-learn openpyxl

# Import pustaka
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from google.colab import files

# Upload file Excel
uploaded = files.upload()
```

```python
# Baca nama file yang diupload
file_name = list(uploaded.keys())[0]

# Baca file Excel
df = pd.read_excel(file_name)

# Tampilkan 5 data teratas
df.head()
```

```python
# Ambil kolom yang sesuai: Ulasan & Sentimen
df = df.iloc[:, :3]  # kolom ke-1 No, ke-2 Ulasan, ke-3 Sentimen
df.columns = ['No', 'Ulasan', 'Sentimen']

# Hapus baris kosong
df.dropna(inplace=True)

# Ubah label sentimen menjadi lowercase (positif / negatif)
df['Sentimen'] = df['Sentimen'].str.lower()

# Lihat jumlah data
df['Sentimen'].value_counts()
```

```python
# Fitur dan Label
X = df['Ulasan']
y = df['Sentimen']

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Train-test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Naïve Bayes Classifier
model = MultinomialNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi Model Naïve Bayes: {:.2f}%".format(accuracy * 100))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```
